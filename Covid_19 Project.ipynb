```python
df_world=df[df['location']=='World']
df_world
```

```python
df_world.reset_index(drop=True)
```

```python
df_world[df_world['new_deaths']==df_world['new_deaths'].max()]['date']
```

```python
df_world_recent=df_world[df_world['date']==df_world['date'].max()]
df_world_recent
```

```python
df_world_recent=df_world_recent[['date','total_cases','new_cases','total_deaths','new_deaths']]
df_world_recent.reset_index(drop=True)
```

```python
df_world['total_cases'].max()
```

```python
df_world['population'].unique
```

```python
df_world['population'].max()
```

```python
df_world_ratio = df_world['total_cases'].max() / df_world['population'].max()
df_world_ratio
```

```python
print('{:.2f}'.format(df_world_ratio))
```

```python
df_world_over_time=df_world.groupby('date')[['date','total_cases','new_cases','total_deaths','new_deaths']].sum().reset_index()
df_world_over_time
```

```python
df_world_over_time.info()
```

```python
df_world_over_time['date']=pd.to_datetime(df_world_over_time['date'])
```

```python
df_world_over_time.info()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'],df_world_over_time['total_cases'])
plt.title('Evolution of Confirmed Covid-19 cases over time in the word',fontsize=16)
plt.xlabel('Days',fontsize=16)
plt.ylabel('Confirmed cases',fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'],df_world_over_time['total_deaths'])
plt.title('Evolution of Covid-19 Deaths cases over time in the world', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('Number of Deaths', fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'],df_world_over_time['new_cases'])
plt.title('Evolution of Covid-19 New Cases over time in the world', fontsize=16)
plt.xlabel('Days', fontsize=16)
plt.ylabel('New Cases', fontsize=16)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time.index, df_world_over_time['total_cases'], label='Confirmed')
plt.plot(df_world_over_time.index, df_world_over_time['total_deaths'], label='Total Deaths')
plt.plot(df_world_over_time.index, df_world_over_time['new_cases'], label='New cases')
plt.plot(df_world_over_time.index, df_world_over_time['new_deaths'], label='New Deaths')
plt.legend()
plt.show()
```
